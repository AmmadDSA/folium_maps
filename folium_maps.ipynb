```python
m2= folium.Map(location=[43.653425095385685, -79.3845425442276], zoom_start=16)
```

```python

folium.Marker([43.65272618249476, -79.38336510390961],
              popup="<h1>Nathan-Phillip Square</h1><img src='conor-samuel.jpg' width=400px><p>Hangout place</p>",
              tootip="The Square", 
              icon= folium.Icon(icon='leaf', icon_color='red', color='green') 
             ).add_to(m2)
m2.save('map.html')


folium.Marker([43.65599791966314, -79.38057673775313],
              popup="<h1>Dundas Square</h1><img src='data/marcin-skalij.jpg' width=400px><p>Hangout 2</p>",
              tootip="The Other Square", 
              icon= folium.Icon(icon='usd', icon_color='red', color='green') 
             ).add_to(m2)
m2.save('map.html')
```
