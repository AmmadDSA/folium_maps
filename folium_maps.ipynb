Create map of traffic-light cameras 
read marker data from csv

data source
https://open.toronto.ca/dataset/traffic-cameras/

```python
tr_map= folium.Map(location=[43.653425095385685, -79.3845425442276], zoom_start=16)
```

```python
# read csv and strip white spaces
camera = pd.read_csv(r'data/traffic_cameras.csv' , sep='\s*,\s*', engine='python')
```

```python
camera.head()
```

```python
for index, row in camera.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']], 
                  popup=row['Main Road'],
                  icon=folium.Icon(icon='camera')
                 ).add_to(tr_map)
```
